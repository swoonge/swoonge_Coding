```python
import networkx as nx
import matplotlib.pyplot as plt

# Your graph represented as an adjacency list
graph = {
    'A': ['C', 'E'],
    'B': ['C', 'D'],
    'C': ['A', 'B', 'D'],
    'D': ['B', 'C'],
    'E': ['A'],
    'F': ['E'],
    'G': ['E'],
    'H': ['F'],
}

# Convert the adjacency list to a NetworkX graph
G = nx.from_dict_of_lists(graph)

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```
