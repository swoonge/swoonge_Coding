11724번: 연결 요소의 개수
- 문제: 방향 없는 그래프가 주어졌을 때, 연결 요소 (Connected Component)의 개수를 구하는 프로그램을 작성하시오.
- 입력: 첫째 줄에 정점의 개수 N과 간선의 개수 M이 주어진다. (1 ≤ N ≤ 1,000, 0 ≤ M ≤ N×(N-1)/2) 둘째 줄부터 M개의 줄에 간선의 양 끝점 u와 v가 주어진다. (1 ≤ u, v ≤ N, u ≠ v) 같은 간선은 한 번만 주어진다.
- 첫째 줄에 연결 요소의 개수를 출력한다.
- 예제 입력
6 5
1 2
2 5
5 1
3 4
4 6
- 예제 출력
2

```python
data = [6,5,1, 2,2, 5,5, 1,3, 4,4, 6]
```

```python
def bfs(sNode, graph, visited):
    queue = []
    queue.append(sNode)
    while queue:
        visited[queue[0]] = True
        for idx, i in enumerate(graph[queue[0]]):
            if (i == 1) and (idx not in queue) and (not visited[idx]):
                queue.append(idx)
        queue = queue[1:]

def makeAdjacencyMatrix(data):
    aMatrix = [[0 for _ in range(data[0]+1)] for _ in range(data[0]+1)]
    for i in range(2, int(len(data)),2):
        aMatrix[data[i]][data[i+1]] = 1
        aMatrix[data[i+1]][data[i]] = 1
    return aMatrix

def main(data):
    count = 0
    aMatrix = makeAdjacencyMatrix(data)
    visited = [False for _ in range(len(aMatrix))]
    visited[0] = True

    while (False in visited):
      bfs(visited.index(False), aMatrix, visited)
      count += 1
      print("fi")

    print(count)

main(data)
    
```

```python
import sys
sys.setrecursionlimit(10**7)
input = sys.stdin.readline

n, l = map(int, input().split())

aMatrix = [[False for _ in range(n+1)] for _ in range(n+1)]
visited = [False] * (n+1)
visited[0] = True

for _ in range(l):
    u, v = map(int, input().split())
    aMatrix[u][v] = True
    aMatrix[v][u] = True
            
def dfs(cNode, visited, aMatrix):
    visited[cNode] = True
    for idx, i in enumerate(aMatrix[cNode]):
        if (i == True) and (not visited[idx]):
            dfs(idx, visited, aMatrix)

count = 0

while (False in visited):
    dfs(visited.index(False), visited, aMatrix)
    count += 1

print(count)
```
